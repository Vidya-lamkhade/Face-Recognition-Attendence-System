```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1)Read the Data:
df=pd.read_csv("StudentDetails.csv")
```

```python
#2)DataFrame Information:
print("DataFrame Info:")
```

```python
print(df.dropna())
```

```python
print(df.info())
```

```python
#3)Display summary statistics
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Check for missing values
print("\nMissing Values:")

print(df.isnull().sum())
```

```python
# Plotting enrollment distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Enrollment'], bins=20, kde=True)
plt.title('Enrollment Distribution')
plt.xlabel('Enrollment')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting enrollment over time
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
plt.figure(figsize=(12, 6))
sns.lineplot(x='DateTime', y='Enrollment', data=df, marker='o')
plt.title('Enrollment Over Time')
plt.xlabel('Date and Time')
plt.ylabel('Enrollment')
plt.xticks(rotation=45)
plt.show()
```

```python
# Display a count plot of names
plt.figure(figsize=(14, 8))
sns.countplot(y='Name', data=df, order=df['Name'].value_counts().index)
plt.title('Count of Enrollments by Name')
plt.xlabel('Count')
plt.ylabel('Name')
plt.show()
```
